```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 f = open("JS04-T004-0001.dat", "r")  ## f = open("filename.dat", "r") f.readline()
```

```python
file_names = ["JS04-T004-0001.dat", "JS04-T004-0002.dat", "JS04-T004-0003.dat", "JS04-T004-0004.dat", "JS04-T004-0005.dat"]
```

```python
f.readline()
```

```python
voltage = []
current = [] 

for file_name in file_names:
    f = open(file_name, "r")
    
    for line in f:
        data = line.split()
        voltage.append(float(data[0]))
        current.append(float(data[1]))   
    plt.plot(voltage, current, 'bo')
    
plt.xlabel("voltage")
plt.ylabel("current")
plt.show()
 
```

```python
voltages = []
currents = [] 

voltage = []
current = [] 

for file_name in file_names:
    f = open(file_name, "r")
    
    for line in f:
        data = line.split()
        voltage.append(float(data[0]))
        current.append(float(data[1]))
    
    voltages.append(voltage)
    currents.append(current)
    
    plt.xlabel("voltage")
    plt.ylabel("current")
    
    plt.plot(voltage, current, 'bo-')
    plt.show()
```
